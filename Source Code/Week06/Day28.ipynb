```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm tính đạo hàm bằng sai phân hữu hạn
def numerical_derivative(f, x, h =1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

# Hàm vẽ đồ thị
def plot_function_and_derivative (f, x_range = (-2, 2), title = "Đồ thị f(x) và f'(x)") :
    x_vals = np.linspace(x_range[0], x_range[1], 400)
    y_vals = f(x_vals)
    y_derivative_vals = numerical_derivative(f, x_vals)

    plt.figure(figsize = (8, 5))
    plt.plot(x_vals, y_vals, label = "f(x)", color = 'blue')
    plt.plot(x_vals, y_derivative_vals, label ="f'(x)(numerical)", color = 'red', linestyle = "--")
    plt.axhline(0, color = 'black', linewidth = 0.5, linestyle = '--')
    plt.axvline(0, color = 'black', linewidth = 0.5, linestyle = '--')
    plt.xlabel("x")
    plt.ylabel("f(x) & f'(x)")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def f_b1(x):
    return x**2 - 3*x + 2

plot_function_and_derivative(f_b1)
```

```python
def f_b2(x):
    return x**3 - 3*x**2 + 4*x - 2

plot_function_and_derivative(f_b2)
```

```python
def f_b3(x):
    return np.exp(x)

plot_function_and_derivative(f_b3)
```

```python
def f_b4(x):
    return np.log(x)

plot_function_and_derivative(f_b4, x_range = (0.1, 10))
```

```python
def f_b5(x):
    return np.sin(x)

plot_function_and_derivative(f_b5)
```

```python
def f_b6(x):
    return np.cos(x)

plot_function_and_derivative(f_b6)
```

```python
def f_b7(x):
    return np.tan(x)

plot_function_and_derivative(f_b7, x_range = (-np.pi/2, np.pi/2))
```

```python
def f_b8(x):
    return np.sqrt(x)

plot_function_and_derivative(f_b8, x_range = (0.1, 10))
```
