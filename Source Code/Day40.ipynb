```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker = 'o', linestyle = '-', color = 'b', label = "Dữ liệu")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Biểu đồ đường với Matplotlib")
plt.legend()
plt.show()
```

#### Bài 2

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 5))

ax1.hist(data["Total Fat"], bins = 15, edgecolor = "black")
ax1.set_xlabel("Total Fat")
ax1.set_ylabel("Count")
ax1.set_xlim(0, 120)

sns.histplot(data["Total Fat"], kde = True, bins = 15, edgecolor = "black", ax = ax2)
ax2.set_xlabel("Total Fat")
ax2.set_ylabel("Density")
ax2.set_xlim(0, 120)

plt.tight_layout()
plt.show()
```

#### Bài 4

```python
from scipy import stats

correlation, p_value = stats.pearsonr(data["Total Fat"], data["Calories"])

plt.figure(figsize = (10, 6))
sns.regplot(x = data["Total Fat"], y = data["Calories"]
            , scatter = True
            , ci = 95
            , line_kws = {"color": "blue"}
            , scatter_kws = {"alpha": 0.5})
plt.xlabel("Total Fat")
plt.ylabel("Calories")

plt.xlim(0, 120)
plt.ylim(0, 2000)

plt.tight_layout()
plt.show()
```

#### Bài 5

```python
chart = sns.jointplot(x = data["Total Fat"], y = data["Calories"]
                      , height = 10
                      , kind = "reg"
                      , marginal_kws = dict(bins = 15)
                      , joint_kws = {"scatter_kws": {"alpha": 0.5}, 
                                     "line_kws": {"color": "blue"}})

chart.set_axis_labels("Total Fat", "Calories")

chart.ax_marg_x.set_xlim(0, 120)
chart.ax_marg_y.set_ylim(0, 2000)

chart.ax_joint.grid(True, alpha = 0.8)

plt.tight_layout()
plt.show()
```

#### Bài 6

```python
plt.figure(figsize=(10, 6))

sns.boxplot(y = data["Sodium"], color = "blue")

plt.title("Biểu đồ Boxplot kiểm tra Outlier của Sodium")
plt.ylabel("Sodium")

plt.tight_layout()
plt.show()
```

```python
data['Sodium'].describe()
```

#### Bài 7

```python
outliers = data[data['Sodium'] > 865 + (1.5*(865-107.5))]['Sodium']
data = data[~data['Sodium'].isin(outliers)]
```

#### Bài 8

```python
category_counts = data.groupby('Category').agg({
    'Item': 'count'
}).reset_index()

category_counts
```

#### Bài 9

```python
plt.figure(figsize=(12, 6))

sns.barplot(x='Category', y='Item'
            , data=category_counts
            , palette='viridis')

plt.title('Biểu đồ số lượng Item có trong mỗi Category')
plt.xlabel('Category')
plt.ylabel('Item Count')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
