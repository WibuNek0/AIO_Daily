```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker = 'o', linestyle = '-', color = 'b', label = "Dữ liệu")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.title("Biểu đồ đường với Matplotlib")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dữ liệu giả lập
data = [10, 20, 25, 30, 40, 35, 50, 60, 55]

sns.histplot(data, kde = True, color = 'blue')
plt.title("Biểu đồ phân bố với Seaborn")
plt.show()
```

```python
import plotly.express as px

df = px.data.iris() # Dữ liệu mẫu

fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species", title="Biểu đồ tán xạ với Plotly")
fig.show()
```
