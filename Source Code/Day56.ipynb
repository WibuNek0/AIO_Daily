```python
import numpy as np

data = np.array([1, 2, 3, 4, 5])

# Tổng thể
std_population = np.std(data)

# Mẫu
std_sample = np.std(data, ddof =1)

print("Population Std:", std_population)
print("Sample Std:", std_sample)
```
